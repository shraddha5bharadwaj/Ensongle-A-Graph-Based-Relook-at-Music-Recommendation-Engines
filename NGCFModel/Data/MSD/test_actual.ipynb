```python
from tqdm import tqdm
import numpy as np
import json
import pandas as pd
```

```python
%cd /content/drive/MyDrive/NGCF_Float64/Data/MSD
```

```python
with open("user_mapping.json") as fp:
    user_mapping = json.load(fp)

with open("song_mapping.json") as fp:
    song_mapping = json.load(fp)

user_inv_mapping = {}
song_inv_mapping = {}
for i in user_mapping:
    user_inv_mapping[str(user_mapping[i])] = i
for i in song_mapping:
    song_inv_mapping[str(song_mapping[i])] = i
print(len(song_inv_mapping), len(user_inv_mapping))
```

```python
f = open("test.txt")
lis = f.readlines()
len(lis)
test_actual = {}
for i in tqdm(lis, ncols=100):
    user_song_lis = i.split()
    # print(user_song_lis)
    if len(user_song_lis) < 2:
        continue
    user = user_inv_mapping[user_song_lis[0]]
    val = []
    for k in user_song_lis[1:]:
        val.append(song_inv_mapping[k])
    test_actual[user] = val
len(test_actual)
with open("test_actual.json", "w") as fp:
    json.dump(test_actual, fp)
```

```python
f = open("train.txt")
lis = f.readlines()
len(lis)
test_actual = {}
for i in tqdm(lis, ncols=100):
    user_song_lis = i.split()
    # print(user_song_lis)
    if len(user_song_lis) < 2:
        continue
    user = user_inv_mapping[user_song_lis[0]]
    val = []
    for k in user_song_lis[1:]:
        val.append(song_inv_mapping[k])
    test_actual[user] = val
len(test_actual)
with open("train_actual.json", "w") as fp:
    json.dump(test_actual, fp)
```
